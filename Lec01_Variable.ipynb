<h1 align="center" style = "color:blue">  Types of variables </h1>

### 1. What is the scope of a variable in Python?

Answer: The scope of a variable defines where in the code a variable is accessible. In Python, variables have either local or global scope. Variables defined within a function have local scope and are only accessible within that function. Variables defined outside of any function have global scope and can be accessed throughout the program.

# i. Local Variable:

A local variable is a variable declared within a function. It has local scope, meaning it is accessible only within that function. It cannot be accessed from outside the function.

```python
def add(): 
    a = 20
    b = 30
    c = a + b
    print('The sum of two num is:', c)
add()
```

## ii. Global Variable:

A global variable is a variable declared outside of any function or block. It has global scope, meaning it can be accessed and modified from any part of the code, including within functions.

- scope inside and outside the functiom. use global() keyword

```python
x = 101               # declare variable
def mainFun():
    global x
    print(x)
    x = 'welocme to python'          #modify global var
    print(x)
```

<h1 align="center" style = "color:blue">  i. Numeric DataType </h1>

Python supports three types of numeric data.

1. **Int** - Integer value can be any length such as integers 10, 2, 29, -20, -150 etc. Python has no restriction on the length of an integer. Its value belongs to int
2. **Float** - Float is used to store floating-point numbers like 1.9, 9.902, 15.2, etc. It is accurate upto 15 decimal points.
3. **complex** - A complex number contains an ordered pair, i.e., x + iy where x and y denote the real and imaginary parts, respectively. The complex numbers like 2.14j, 2.0 + 2.3j, etc.
4. **Boolean:** Also known as bool data type allows us to choose between two values: True and False. Booleans have the type bool. A Boolean is used to determine whether the logic of an expression or a comparison is correct. It plays a huge role in data comparisons.

## 1. Implicit Type Conversion:

- In Implicit Type Conversion, python automatically converts one data type into another data type. 

- While performing arithmetic operations, integers are automatically converted to float if any of the operands is a float. Also, the division operator / always returns a float, even if both operands are integers. Use the // operator if you want the result of the division to be an int.

```python
a = 7
b = 8
value = a/b
print(type(value))
```

```python
x = 10
y = 5.5
total = 10 + 5
print(total)
print(type(total))
```

```python
a = 2 + 3
b = 2 + 3.0
type(a), type(b)
```

```python
a = 4/2
b = 4/2.0
print(type(a), type(b))
```

```python
a = 3 * 5
b = 3 * 5.0
type(a), type(b)
```

```python
age = 21
marks = 74.2
c = age + marks
print(c)
print(type(c))
```

```python
number1 = 63
number2 = 8
print(number1/number2)
print(number1//number2)          # floor division

print(type(number1/number2))
print(type(number1//number2))
```

## 2. Explicit Type Conversion:

- or Type Casting refers to the conversion of an object from one data type to another by programmer.
